```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

%matplotlib inline
```

### Pie Charts

```python
days = np.array([1,2,3,4,5])
# 24 hr days
sleeping = [7,8,6,11,7]
eating = [2,3,4,3,2]
working = [7,8,7,2,2]
playing = [8,5,7,8,13]
```

```python
slices = [7,2,2,13]
activities = ['sleeping','eating','working','playing']
colors = ['b','g','r','c']
```

```python
plt.figure(figsize = (10,10))

plt.pie(slices, 
        labels = activities, 
        colors = colors, 
        startangle = 90, 
        shadow = True, 
        explode = (.1,0,0,0),
        autopct = '%1.1f%%'
       )

#plt.xlabel('x',fontsize = 15)
#plt.ylabel('y',fontsize = 15)
plt.title('Interesting Pie Chart',fontsize = 20)
#plt.tick_params(axis = 'both',labelsize = 15)
#plt.legend(loc = 'lower left',fontsize = 15)

plt.show()
```
