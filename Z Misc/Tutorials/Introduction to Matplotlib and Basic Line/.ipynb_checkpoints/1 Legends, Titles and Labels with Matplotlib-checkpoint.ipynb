```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

%matplotlib inline
```

```python
x = pd.Series([0,1,2,3,4,5])
y = pd.Series([0,1,2,3,4,5])
```

```python
plt.figure(figsize = (15,5))

plt.plot(x,y, label = 'First line')
plt.plot(x,y+4,label = 'Second line')

plt.xlabel('x values',fontsize = 20)
plt.ylabel('y values',fontsize = 20)
plt.title('Interesting graph\n check it out',fontsize = 30)
plt.tick_params(axis = 'both',which = 'major',labelsize = 14)
plt.legend(loc = 'center',fontsize = 20)

plt.show()
```
