```python
# Libraries
#---Pandas
import pandas as pd
from pandas.tools.plotting import scatter_matrix

#---numpy
import numpy as np

#---matplotlib
from matplotlib import pyplot as plt
```

```python
# importing
xls = pd.ExcelFile("data/Data_Cortex_Nuclear.xls")
xls.sheet_names
```

```python
# creating a dataframe
data = xls.parse("Hoja1")
len(data.index)
```

```python
# looping over all the indicies with 18 or less missing values...
for x in data.columns[1:44]:
    #replacing null values with mean.
    data[x].fillna(data[x].mean(),inplace = True)
```

```python
"""


This is more of a look into the other expressions so we can figure out the proteins expressed more 
than the others and focus on those.



"""
```
