```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

%matplotlib inline
```

```python
x = pd.Series([2,4,6,8,10])
y = pd.Series([6,7,8,2,4])

x2 = np.array([1,3,5,9,11])
y2 = np.array([7,8,2,4,2])
```

### Bar Plots

```python
plt.figure(figsize = (15,5))

plt.bar(x,y,label = 'bar 1', color = 'red')

plt.xlabel('x values',fontsize = 20)
plt.ylabel('y values',fontsize = 20)
plt.title('Interesting graph\n check it out',fontsize = 30)
plt.tick_params(axis = 'both',labelsize = 14)
plt.legend(loc = 'upper right',fontsize = 20)

plt.show()
```

```python
plt.figure(figsize = (15,5))

plt.bar(x , y,label = 'bar 2', color = 'green')
plt.bar(x2,y2,label = 'bar 1', color = 'blue')

plt.xlabel('x values',fontsize = 20)
plt.ylabel('y values',fontsize = 20)
plt.title('Interesting graph\n check it out',fontsize = 30)
plt.tick_params(axis = 'both',which = 'major',labelsize = 14)
plt.legend(loc = 'upper right',fontsize = 20)

plt.show()
```

### Histograms

```python
population_ages = pd.Series([22,55,62,45,21,22,34,42,42,4,99,102,120,121,122,130,111,115,112,80,75,65,54,44,43,42,48])
bins = [x for x in range(0,140,10)]
```

```python
plt.figure(figsize = (15,5))

plt.hist(population_ages,bins,histtype = 'bar', alpha = .5, rwidth = 0.8,label = 'hist 1')

plt.xlabel("ages",fontsize = 15)
plt.ylabel('counts in bin',fontsize = 15)
plt.tick_params(axis = 'both',labelsize = 13)
plt.legend(loc = 'upper right')

plt.show()
```
