```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

%matplotlib inline
```

### Scatter Plots

```python
x = pd.Series(range(1,9))
y = pd.Series([5,2,4,2,1,4,5,2])
```

```python
plt.figure(figsize = (15,5))

plt.scatter(x,y,label='scatter 1',color = 'g', marker = 'x', s = 100)

plt.xlabel("x values",fontsize = 15)
plt.ylabel("y values",fontsize = 15)
plt.title("Interesting Graph",fontsize = 20)
plt.tick_params(axis = 'both', labelsize = 15)
plt.legend(loc = 'lower left')

plt.show()
```
