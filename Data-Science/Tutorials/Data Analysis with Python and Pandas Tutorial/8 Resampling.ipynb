```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import quandl

import pickle

from matplotlib import style
style.use('ggplot')
```

```python
HPI_data = pd.read_pickle('data/new_file3.pickle')
HPI_data.head()
```

```python
plt.figure()
ax1 = plt.subplot2grid((1,1),(0,0))

#resample
"""
Alias	Description
B	business day frequency
C	custom business day frequency (experimental)
D	calendar day frequency
W	weekly frequency
M	month end frequency
BM	business month end frequency
CBM	custom business month end frequency
MS	month start frequency
BMS	business month start frequency
CBMS	custom business month start frequency
Q	quarter end frequency
BQ	business quarter endfrequency
QS	quarter start frequency
BQS	business quarter start frequency
A	year end frequency
BA	business year end frequency
AS	year start frequency
BAS	business year start frequency
BH	business hour frequency
H	hourly frequency
T	minutely frequency
S	secondly frequency
L	milliseonds
U	microseconds
N	nanoseconds
"""
CA1yr = HPI_data['ValueCA'].resample('A',how = 'mean')
print CA1yr.head()

HPI_data['ValueCA'].plot(ax=ax1,color = 'b',label = 'default: monthly')
CA1yr.plot(ax = ax1,color = 'g',label = 'resampled: yearly')

plt.legend(loc = 'upper left')
plt.show()
```

```python
plt.figure()
ax1 = plt.subplot2grid((1,1),(0,0))

CA1yr = HPI_data['ValueCA'].resample('A',how = 'ohlc')
print CA1yr.head()

HPI_data['ValueCA'].plot(ax=ax1,color = 'b',label = 'default: monthly')
CA1yr.plot(ax = ax1,label = 'resampled: ohlc')

plt.legend(loc = 'upper left')
plt.show()
```
