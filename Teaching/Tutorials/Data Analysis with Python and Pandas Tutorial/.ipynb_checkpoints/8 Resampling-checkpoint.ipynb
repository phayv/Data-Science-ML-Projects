### What tells us more is percent change

```python
api_key = 'oaoRfxFc9iKgnqsudsMa'
```

```python
def state_list():
    return pd.read_excel('data/fifty-states.xlsx',header=None)[0]
```

```python
def grab_initial_state_data():
    states_list = state_list()
    main_df = pd.DataFrame()
    
    for states in states_list:
        query = 'FMAC/HPI_'+str(states)
        df = quandl.get(query, authtoken = api_key)
        
        # NEW
        df = df.pct_change()
    
        if main_df.empty:
            main_df = df
        else:
            #indexed by date, so join is the best because they all have the same dates
            main_df = main_df.join(df,lsuffix=states)
    
    # makes a new file or open existing (# NEW FILE )
    pickle_out = open('data/new_file2.pickle','wb')
    # putting the dataframe in
    pickle.dump(main_df,pickle_out)
    # closing it because we opened it
    pickle_out.close()
```

```python
def grab_initial_state_data():
    states_list = state_list()
    main_df = pd.DataFrame()
    
    for states in states_list:
        query = 'FMAC/HPI_'+str(states)
        df = quandl.get(query, authtoken = api_key)

        # NEW
        df['Value'] = (df['Value'] - df['Value'][0]) / df['Value'][0] * 100
    
        if main_df.empty:
            main_df = df
        else:
            #indexed by date, so join is the best because they all have the same dates
            main_df = main_df.join(df,lsuffix=states)
    
    # makes a new file or open existing (# NEW FILE )
    pickle_out = open('data/new_file3.pickle','wb')
    # putting the dataframe in
    pickle.dump(main_df,pickle_out)
    # closing it because we opened it
    pickle_out.close()
```

```python
 grab_initial_state_data()
```

```python
HPI_data = pd.read_pickle('data/new_file3.pickle')
HPI_data.head()
```

```python
HPI_data.plot()
plt.legend().remove()
plt.show()
```

### Adding Entire USA

```python
def HPI_Benchmark():
    df = quandl.get('FMAC/HPI_USA', authtoken = api_key)
    df['Value'] = (df['Value'] - df['Value'][0]) / df['Value'][0] * 100
    return df
```

```python
HPI_USA = HPI_Benchmark()
```

```python
fig = plt.figure()
ax1 = plt.subplot2grid((1,1),(0,0))

HPI_data.plot(ax = ax1)
HPI_USA.plot(ax = ax1,color='k',linewidth = 10)
plt.legend().remove()
plt.show()
```
