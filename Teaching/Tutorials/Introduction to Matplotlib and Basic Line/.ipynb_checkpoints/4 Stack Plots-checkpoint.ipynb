```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

%matplotlib inline
```

### Stack Plots

```python
days = np.array([1,2,3,4,5])
# 24 hr days
sleeping = [7,8,6,11,7]
eating = [2,3,4,3,2]
working = [7,8,7,2,2]
playing = [8,5,7,8,13]
```

```python
plt.figure(figsize = (15,5))

plt.plot([],[],color = 'c',label = 'playing',linewidth = 5)
plt.plot([],[],color = 'r',label = 'working',linewidth = 5)
plt.plot([],[],color = 'g',label = 'eating',linewidth = 5)
plt.plot([],[],color = 'b',label = 'sleeping',linewidth = 5)

plt.stackplot(days, sleeping,eating,working,playing, colors = ['b','g','r','c'])

plt.xlabel("x values",fontsize = 14)
plt.ylabel("y values",fontsize = 14)
plt.title("interesting graph",fontsize = 14)
plt.tick_params(axis = 'both',labelsize = 15)
plt.legend(loc = 'upper left',fontsize = 15)

plt.show()
```
