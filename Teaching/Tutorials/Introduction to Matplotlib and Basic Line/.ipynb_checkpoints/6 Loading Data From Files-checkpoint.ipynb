```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import csv

%matplotlib inline
```

## Loading Data From Files

### First Way

```python
x = []
y = []

with open('data/test.csv','r') as csvfile:
    plots = csv.reader(csvfile,delimiter = ',')
    for row in plots:
        x.append(int(row[0]))
        y.append(int(row[1]))
```

```python
plt.figure(figsize = (15,5))

plt.plot(x,y,label = 'loaded from file')

plt.xlabel('x')
plt.ylabel('y')
plt.title('interesting plot')
plt.tick_params()
plt.legend(loc = 'upper left')

plt.show()
```

### Second Way

```python
x,y = np.loadtxt('data/test.csv',delimiter=',',unpack =True)
```

```python
plt.figure(figsize = (15,5))

plt.plot(x,y,label = 'loaded from file')

plt.xlabel('x')
plt.ylabel('y')
plt.title('interesting plot')
plt.tick_params()
plt.legend(loc = 'upper left')

plt.show()
```
