```python
import matplotlib.pyplot as plt
from matplotlib import style
import numpy as np
from sklearn.cluster import KMeans
style.use('ggplot')

%matplotlib inline
```

```python
X = np.array([[1, 2],
              [1.5, 1.8],
              [5, 8],
              [8, 8],
              [1, 0.6],
              [9, 11]])

plt.figure(figsize = (15,10))
plt.scatter(X[:, 0],X[:, 1], s=150, linewidths = 5, zorder = 10)
plt.show()
```

```python
#default is 8

KM = KMeans(n_clusters = 2)
KM.fit(X)

centroids = KM.cluster_centers_
labels = KM.labels_

colors = ['b+','g+']
```

```python
plt.figure(figsize = (15,5))

for i in range(len(X)):
    plt.plot(X[i][0],X[i][1], colors[labels[i]], markersize = 20)
    
plt.scatter(centroids[:,0], centroids[:,1], marker = 'x', s = 100, linewidth = 5)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
import seaborn as sns
```

```python
# Search for cluster center (mean of points)
from sklearn.datasets.samples_generator import make_blobs

x,y = make_blobs(n_samples = 500, centers = 5, random_state = 0, cluster_std = .7)

plt.figure(figsize = (15,10))
plt.scatter(x[:,0],x[:,1], s = 30)
plt.show()
```

```python
# one set is easier to see, but we don't know where one ends and the other begins in the other 4
from sklearn.cluster import KMeans

exp = KMeans(5)
exp.fit(x)
y_means = exp.predict(x)

plt.figure(figsize = (15,10))
plt.scatter(x[:,0],x[:,1], c = y_means, s = 30, cmap = 'rainbow')
plt.show()
print "--- The algorithm knows better and follows what we would do by eye in terms of grouping."
```
